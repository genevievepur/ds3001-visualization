##3. Cleaning Variables ##
**Pick out a few interesting variables, and explain why you find them jointly interesting. Clean those variables, documenting your choices by commenting in the code or in a markdown chunk in a notebook.**

```python
pd.set_option('display.float_format', '{:.0f}'.format)
df = pd.read_csv('/CostReport_2021_Final.csv',low_memory=False)
print(df.shape, '\n')
print(df.columns, '\n')
df.head()
```

```python
df = df.rename(columns = {'Rural Versus Urban' : 'area_type', 'Provider Type' : 'provider_type',
                          'Type of Control' : 'control_type', 'Number of Beds' : 'num_beds',
                          'Cost of Charity Care' : 'charity_care_cost', 'Cost of Uncompensated Care' : 'uncomp_care_cost'})
print(df.columns, '\n')
```

```python
# RURAL VERSUS URBAN
area_type = df['area_type']
print(area_type.unique(), '\n')
print(area_type.value_counts(), '\n')
print('Total missing: ', sum(area_type.isnull()))
```

```python
# PROVIDER TYPE
provider = df['provider_type']
print(provider.unique(), '\n')
print(provider.value_counts(), '\n')
print('Total missing: ', sum(provider.isnull()))
```

```python
# TYPE OF CONTROL
control = df['control_type']
print(control.unique(), '\n')
print(control.value_counts(), '\n')
print('Total missing: ', sum(control.isnull()))
```

For the categorical variables, they came in already clean so I did not wrangle them any further. There were no categories that needed to be grouped together or transformed into NaNs.

```python
# NUMBER OF BEDS
beds = df['num_beds']
print(beds.unique(), '\n')
print(beds.describe(), '\n')
print('Total missing: ', sum(beds.isnull()))
```

Because the values appear in standard deviation as data type float64, I wanted to convert them to int64s so that they are easier to view and understand. To do this, I first created a missing value dummy variable and infinite value dummy variable. Then, imputed the missing and infinite values with the mean to keep the statistics the same. I had to use an impute value because I as not able to convert to an int64 with NaNs and infs. Thus, I was able to convert the values to integers.

```python
# CLEANING NUMBER OF BEDS
df['beds_nan'] = beds.isnull()
df['beds_inf'] = beds.isin([np.inf, -np.inf])
impute_value = beds.mean()
beds = beds.fillna(impute_value)
beds.replace([np.inf, -np.inf], impute_value, inplace=True)
beds = beds.astype('int64')
print(beds.unique(), '\n')
print(beds.describe(), '\n')
print('Total missing: ', sum(beds.isnull()))
```

```python
# COST OF CHARITY CARE
charity = df['charity_care_cost']
print(charity.unique(), '\n')
print(charity.describe(), '\n')
print('Total missing: ', sum(charity.isnull()))
```

Because the format of the values is similar to how they were in 'Number of Beds', I performed the same process. Additionally, I replaced the NaNs and infs with the mean of the dataset.

```python
# CLEANING COST OF CHARITY CARE
df['charity_nan'] = charity.isnull()
df['charity_inf'] = charity.isin([np.inf, -np.inf])
impute_value = charity.mean()
charity = charity.fillna(impute_value)
charity.replace([np.inf, -np.inf], impute_value, inplace=True)
charity = charity.astype('int64')
print(charity.unique(), '\n')
print(charity.describe(), '\n')
print('Total missing: ', sum(charity.isnull()))
```

```python
# COST OF UNCOMPENSATED CARE
uncomp = df['uncomp_care_cost']
print(uncomp.unique(), '\n')
print(uncomp.describe(), '\n')
print('Total missing: ', sum(uncomp.isnull()))
```

I cleaned the Cost of Uncompensated Care with the same methods as the previous two numerical variables, as the format for this variable was the same.

```python
# CLEANING THE COST OF UNCOMPENSATED CARE
df['uncomp_nan'] = uncomp.isnull()
df['uncomp_inf'] = uncomp.isin([np.inf, -np.inf])
impute_value = uncomp.mean()
uncomp = uncomp.fillna(impute_value)
uncomp.replace([np.inf, -np.inf], impute_value, inplace=True)
uncomp = uncomp.astype('int64')
print(uncomp.unique(), '\n')
print(uncomp.describe(), '\n')
print('Total missing: ', sum(uncomp.isnull()))
```
